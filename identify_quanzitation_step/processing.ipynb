Xac dinh loi luong tu cua tin hieu hinh sin

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
N = 200
n = np.arange(N)
f0 = 1/50
x_n = np.sin(2*np.pi*f0*n)
x_n
```

```python
max_val = np.max(x_n)
min_val = np.min(x_n)
step64 = (max_val-min_val)/(64-1)
step128 = (max_val-min_val)/(128-1)
step256 = (max_val-min_val)/(256-1)
print(step64, step128, step256)
max_val
```

Xac dinh tin hieu sau luong tu xq_n

```python
def determine_xq_n(step):
    xq_n = np.round((x_n-min_val)/step) *step + min_val
    return xq_n
```

```python
xq_n64 = determine_xq_n(step=step64)
xq_n64
```

```python
xq_n128 = determine_xq_n(step=step128)
xq_n128
```

```python
xq_n256 = determine_xq_n(step=step256)
xq_n256
```

Tinh e(n)

```python
e_n128 = xq_n128 - x_n
e_n64 = xq_n64 - x_n
e_n256 = xq_n256 - x_n
```

Tinh SQNR va so sanh voi ly thuyet

```python
SNQR64 = max_val**2 / (np.sum(e_n64 * e_n64)/200 * 2)
SNQR128 = max_val**2 / ( 2* np.sum(e_n128 * e_n128)/200)
SNQR256 = max_val**2 / (2 * np.sum(e_n256 * e_n256)/200)
print(SNQR64, SNQR128, SNQR256, sep='\n')
SNQR64_lt = 3/2 * 2**(2*6)
SNQR128_lt = 3/2 * 2**(2*7)
SNQR256_lt = 3/2 * 2**(2*8)
print(SNQR64_lt, SNQR128_lt, SNQR256_lt)
```

```python
#Doi sang db
SNQR64db = 10* np.log10(SNQR64)
SNQR64_ltdb = 1.76 + 6.02*6
print(SNQR64db, SNQR64_ltdb)
```

```python
#Doi sang db
SNQR128db = 10* np.log10(SNQR128)
SNQR128_ltdb = 1.76 + 6.02*7
print(SNQR128db, SNQR128_ltdb)
```

```python
#Doi sang db
SNQR256db = 10* np.log10(SNQR256)
SNQR256_ltdb = 1.76 + 6.02*8
print(SNQR256db, SNQR256_ltdb)
```

Ve cac tin hieu xq(n) va x(n) tren 1 do thi

```python
plt.plot(n, x_n, label = "x(n)")
plt.plot(n, xq_n64,"o", label = "xq(n)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(n, x_n, label = "x(n)")
plt.plot(n, xq_n64,"o", label = "xq(n)")
ticks = np.linspace(min_val, max_val, 64)
plt.legend()
plt.yticks(ticks=ticks)
plt.ylim(-0.4, 0.5)
plt.xlim(71, 79)
plt.grid(True)
plt.show()
```

```python
plt.plot(n, x_n, label = "x(n)")
plt.plot(n, xq_n128,"o", label = "xq(n)")
ticks = np.linspace(min_val, max_val, 128)
plt.legend()
plt.yticks(ticks=ticks)
plt.ylim(-0.18, 0.26)
plt.xlim(72.5, 76.5)
plt.grid(True)
plt.show()
```

```python
plt.plot(n, x_n, label = "x(n)")
plt.plot(n, xq_n256,"o", label = "xq(n)", ms = 5)
ticks = np.linspace(min_val, max_val, 256)
plt.legend()
plt.yticks(ticks=ticks)
plt.ylim(-0.14, 0.02)
plt.xlim(74.5, 76.5)
plt.grid(True)
plt.show()
```
